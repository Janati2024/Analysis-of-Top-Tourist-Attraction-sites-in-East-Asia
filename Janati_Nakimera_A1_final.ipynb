# Cleaning the data
*   Replace NAN values with zero for numerical columns and unknown for categorical columns
*   Produce duplicate values if any

*   Show recommended plots











```python
# prompt: Replace NAN values with zero for numerical columns and unknown for categorical columns

for col in df_travels.columns:
    if pd.api.types.is_numeric_dtype(df_travels[col]):
        df_travels[col] = df_travels[col].fillna(0)
    else:
        df_travels[col] = df_travels[col].fillna('unknown')
```

# Data Analysis
*  Distribution of reviews across all columns
*  Outliers in ratings
*  Correlation analysis
*  Top attractions by ratings





```python
# Distribution of reviews across all columns
df_travels.hist(figsize=(15, 10), bins=30)
plt.suptitle('Distribution of Reviews')
plt.show()
```

The more spread the values are from the mean implies a larger standard deviation, meaning the data points are distributed further away from the central tendency (mean), indicating greater variability within the dataset. Greater variability is seen in the charts above.

```python
# Boxplot to identify outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_travels, palette='Set2')
plt.title('Boxplot of Ratings Across Attractions')
plt.xticks(rotation=45)
plt.show()
```

```python
# Select only numeric columns for correlation
numeric_columns = df_travels.select_dtypes(include=['float64', 'int64'])

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_columns.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Ensure only numeric columns are used for calculating averages
numeric_columns = df_travels.select_dtypes(include=['float64', 'int64'])

# Check if there are valid numeric columns
if not numeric_columns.empty:
    # Calculate average ratings for each numeric column
    average_ratings = numeric_columns.mean().sort_values(ascending=False)

    # Bar chart for average ratings
    plt.figure(figsize=(12, 6))
    sns.barplot(x=average_ratings.index, y=average_ratings.values, palette='viridis')
    plt.title('Average Ratings of Attractions')
    plt.xticks(rotation=45)
    plt.ylabel('Average Rating')
    plt.xlabel('Attractions')
    plt.show()
else:
    print("No numeric columns available for analysis.")
```

For this part of the analysis, average ratings were computed using only numeric columns. The non numeric ones were left out.This was the same case for getting the correlation values from the heat map.

Based on the findings of this analysis, here are actionable recommendations that can be implemented to optimize operations and improve visitor satisfaction:(OpenAI, 2024)

Enhance Marketing for Top Attractions:

Data-Driven Decisions:

Regularly update and analyze visitor feedback to track changes in satisfaction over time. Introduce tools like visitor surveys or digital feedback systems to complement existing data and provide richer insights for decision-making.
